```python
#Crear un dataframe "datau" que contendra las 100 ultimas filas de 3 columnas:"location",  "date", "new_cases_smoothed"
datau = data.tail(100)[["location", "date", "new_cases_smoothed"]]
#Reinicia el indice de la nueva "datau"
datau.reset_index(drop=True, inplace=True)
#Este es el codigo para que nuestro index de "datau" sea desde el 1 al 100
datau.index += 1
#Activamos la visualizacion total de los registros de un dataframe
pd.set_option('display.max_rows', None)
datau
```

```python
#Solo filtramos la data de PERU y considerar dos columnas de 3 columnas:"location",  "date", "new_cases_smoothed"
#creamos un nuevo data frame con solo registros de Peru
data_peru = data[data["location"] == "Peru"]
#Filtramos solo las columnas que nos interesan
data_peru = data_peru[["location", "date", "new_cases_smoothed"]]
#Activamos la visualizacion total de los registros de un dataframe
pd.set_option('display.max_rows', 10)
data_peru
```

```python
#Nos piden los datos de agosto del 2020 en Perú
#Convertimos la columna 'date' a tipo datetime para facilitar la manipulación de fechas
data_peru['date'] = pd.to_datetime(data_peru['date'])

#Filtramos los registros de agosto de 2020
data_peru_agosto_2020 = data_peru[(data_peru['date'].dt.year == 2020) & (data_peru['date'].dt.month == 8)]

data_peru_agosto_2020
```

```python
#Creamos un grafico de barras de los casos en Peru en el mes de Agosto
data_peru_agosto_2020.plot(x='date', y='new_cases_smoothed', kind='line')
```

```python
#Filtramos datos de los paises andinos (Perú, Argentina, Bolivia y Chile)
data_andinos_dpn = data[(data["location"] == "Peru")|(data["location"] == "Argentina")|(data["location"] == "Bolivia")|(data["location"] == "Chile")|(data["location"] == "Venezuela")]
data_andinos_dpn
```

```python
data_andinos_dpn = data_andinos_dpn[["location", "date", "new_cases_smoothed"]]
data_andinos_dpn
```

```python
#Primero hacemos el arreglo de la columna que tiene registro como Naan lo eliminamos
data_andinos_dpn = data_andinos_dpn.dropna()
data_andinos_dpn
```

```python
#Ahora filtramos la cantidad total de casos en cada pais andino: Perú, Argentina, Bolivia y Chile
#Agrupar por país y sumar los casos
total_data_andinos_dpn = data_andinos_dpn.groupby("location")["new_cases_smoothed"].sum()
#Desactivamos notación científica para pandas
pd.set_option('display.float_format', '{:,.0f}'.format)
total_data_andinos_dpn
```

```python
#Vamos a crear una tabla donde tendremos las columnas: "Años", "Perú", "Argentina", "Chile", "Bolivia" y las filas: "2020", "2021", "2022", "2023", "2024"
#y el total de la suma de casos por año de cada uno de los paises
# Convertimos la columna 'date' a tipo datetime para facilitar la manipulación de fechas
data_andinos_dpn['date'] = pd.to_datetime(data_andinos_dpn['date'])
# Extraemos el año de la columna 'date'
data_andinos_dpn['year'] = data_andinos_dpn['date'].dt.year

```

```python
#Realizamos la suma total de casos por año de Perú
total_peru = data_andinos_dpn[data_andinos_dpn['location'] == 'Peru'].groupby('year')['new_cases_smoothed'].sum()
total_peru
```

```python
import matplotlib.pyplot as plt

# Create the bar plot
plt.bar(total_peru.index, total_peru.values)

# Set the labels and title
plt.xlabel("Years")
plt.ylabel("Cases")
plt.title("Total Cases by Year in Peru")

# Show the plot
plt.show()
```

```python
#Realizamos la suma total de casos por año de Argentina
total_argentina = data_andinos_dpn[data_andinos_dpn['location'] == 'Argentina'].groupby('year')['new_cases_smoothed'].sum()
total_argentina
```

```python
#Realizamos la suma total de casos por año de Chile
total_chile = data_andinos_dpn[data_andinos_dpn['location'] == 'Chile'].groupby('year')['new_cases_smoothed'].sum()
total_chile
```

```python
#Realizamos la suma total de casos por año de Bolivia
total_bolivia = data_andinos_dpn[data_andinos_dpn['location'] == 'Bolivia'].groupby('year')['new_cases_smoothed'].sum()
total_bolivia
```

```python
#Realizamos la suma total de casos por año de Venezuela
total_venezuela = data_andinos_dpn[data_andinos_dpn['location'] == 'Venezuela'].groupby('year')['new_cases_smoothed'].sum()
total_venezuela
```

```python
#Ahora mostramos en una tabla todos los resultados de cada año por pais donde el index sera el año
# Crear un DataFrame con los totales anuales por país
totales_por_pais = pd.DataFrame({
    'Perú': total_peru,
    'Argentina': total_argentina,
    'Chile': total_chile,
    'Bolivia': total_bolivia,
    'Venezuela': total_venezuela
})

# Agregar una fila con la suma total de casos por país
totales_por_pais.loc['Total'] = totales_por_pais.sum()

totales_por_pais
```

```python
#Importamos la libreria para crear un grafico de carrera
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from moviepy.editor import VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage
```

```python
#Creamos gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
totales_por_pais.drop(index='Total').plot(kind='bar', ax=ax)
ax.set_title('Total de casos por país y año')
ax.set_xlabel('Año')
ax.set_ylabel('Total de casos')
plt.tight_layout()
plt.savefig('bar_chart.png')
```

```python
def draw_barchart(year):
    dff = totales_por_pais.drop(index='Total').loc[year]
    ax.clear()
    dff.plot(kind='bar', ax=ax, color=['#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#e4bb81'])
    for i, (value, name) in enumerate(zip(dff.values, dff.index)):
        ax.text(i, value + 100, f'{value:,.0f}', size=12, ha='center')
    ax.set_title(f'Total de casos en {year}')
    ax.set_xlabel('País')
    ax.set_ylabel('Total de casos')
    ax.set_ylim(0, totales_por_pais.drop(index='Total').values.max() * 1.2)
    ax.grid(True, axis='y', linestyle='--', alpha=0.7)
    return mplfig_to_npimage(fig)

# Crear figura y ejes para el gráfico de carrera
fig, ax = plt.subplots(figsize=(10, 6))

# Función para generar cada frame de la animación
def make_frame(t):
    year = totales_por_pais.drop(index='Total').index[int(t)]
    return draw_barchart(year)

# Duración de la animación
duration = len(totales_por_pais.drop(index='Total'))

# Crear la animación
animation = VideoClip(lambda t: make_frame(t), duration=duration)

# Guardar la animación como un archivo mp4
animation.write_videofile("Grafico_carrera_casos.mp4", fps=1)
```
