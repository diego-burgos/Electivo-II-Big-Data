```python
import pandas as pd
```

```python
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSzJDRWDL-gjBRgJaeeNRuzZL7742CR_wLBb_fF-GRe3kpO7EiOEdBtx-j6M2ZkeyQ_deTINX_G5cFw/pub?gid=1188755150&single=true&output=csv")
```

```python
datos
```

```python
#Obtenemos la suma total de la columna ingresos
suma_ingresos = datos['ingresos'].sum()
print("La suma de la columna de ingresos es: ", suma_ingresos)
```

```python
#Promedio de las edades
promedio_edad = datos['edad'].mean()
print("El promedio de edad es: ",int(promedio_edad))
```

```python
# Filtrar por personas de Arequipa
personas_arequipa = datos[datos['ciudad'] == 'Arequipa']

# Seleccionar solo las columnas 'nombre' y 'ciudad'
personas_arequipa = personas_arequipa[['ciudad', 'nombre']]

print("Personas de Arequipa:")
print(personas_arequipa)
```

```python
#Promedio de ingresos por ciudad
promedio_ingresos_por_ciudad = datos.groupby('ciudad')['ingresos'].mean().astype(int);

print("Promedio de ingresos por ciudad:")
print(promedio_ingresos_por_ciudad)
```

```python
#Filtro por persona mas joven y mas mayor
nombre_persona_menor_edad = datos.loc[datos['edad'].idxmin(), 'nombre']
nombre_persona_mayor_edad = datos.loc[datos['edad'].idxmax(), 'nombre']

print("Nombre de la persona más joven:", nombre_persona_menor_edad)
print("Nombre de la persona más mayor:", nombre_persona_mayor_edad)
```

```python
indice_persona_menor_edad = datos['edad'].idxmin()
indice_persona_mayor_edad = datos['edad'].idxmax()

persona_menor_edad = datos.loc[indice_persona_menor_edad]
persona_mayor_edad = datos.loc[indice_persona_mayor_edad]

print("Persona más joven:")
print(persona_menor_edad)

print("\nPersona más mayor:")
print(persona_mayor_edad)
```

```python
#Conteo cuántas personas trabajan por ciudad
conteo_trabajadores = datos.groupby('ciudad')['nombre'].count()

print("Conteo de personas que trabajan por ciudad:")
print(conteo_trabajadores)
```

```python
#opcion 2 conteo por personas que trabajan por ciudad
import numpy as np
datos['trabaja'] = np.random.choice([True, False], len(datos))

# Contar cuántas personas trabajan por ciudad
conteo_trabajadores_por_ciudad = datos.groupby('ciudad').sum()

print("Conteo de personas que trabajan por ciudad:")
print(conteo_trabajadores_por_ciudad)
```

```python
datos
```

```python
#filtro de edad de mayor a menor
df_ordenado_por_edad = datos.sort_values(by='edad', ascending=False)

print("DataFrame ordenado por edad:")
print(df_ordenado_por_edad)
```

```python
# Filtrar las personas mayores de 30 años
mayores_de_30 = datos[datos['edad'] > 30]

# Calcular el promedio de ingresos de las personas mayores de 30 años
promedio_ingresos_mayores_30 = mayores_de_30['ingresos'].mean().astype(int)

print("Promedio de ingresos de personas mayores de 30 años:", promedio_ingresos_mayores_30)
```

```python
#Ingresos totales por ciudad
ingresos_totales_por_ciudad = datos.groupby('ciudad')['ingresos'].sum()

print("Ingresos totales por ciudad:")
print(ingresos_totales_por_ciudad)
```

```python
# Obtener el índice de la persona con el ingreso más alto
indice_persona_mayor_ingreso = datos['ingresos'].idxmax()

# Obtener la persona con el ingreso más alto
persona_mayor_ingreso = datos.loc[indice_persona_mayor_ingreso]

# Imprimir la persona con el ingreso más alto
print("Persona con el ingreso más alto:")
print(persona_mayor_ingreso)
```

```python
#crear histograma de edades

import matplotlib.pyplot as plt

# Crear un histograma de las edades
plt.hist(datos['edad'], bins=10)

# Establecer el título y las etiquetas de los ejes
plt.title('Histograma de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()
```

```python
#agregar una columna de impuestos que represente el 10% de los ingresos

# Agregar una nueva columna llamada 'impuestos'
datos['impuestos'] = datos['ingresos'] * 0.1

# Imprimir el DataFrame actualizado
print(datos)
```
