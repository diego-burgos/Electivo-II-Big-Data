```python
# Datos de conexiones de fibra óptica y convencionales
suma_conexiones_fibra = [
    suma_conexiones_f_1, suma_conexiones_f_2, suma_conexiones_f_3, suma_conexiones_f_4,
    suma_conexiones_f_5, suma_conexiones_f_6, suma_conexiones_f_7, suma_conexiones_f_8,
    suma_conexiones_f_9, suma_conexiones_f_10, suma_conexiones_f_11, suma_conexiones_f_12,
    suma_conexiones_f_13
]

suma_conexiones_convencionales = [
    suma_conexiones_c_1, suma_conexiones_c_2, suma_conexiones_c_3, suma_conexiones_c_4,
    suma_conexiones_c_5, suma_conexiones_c_6, suma_conexiones_c_7, suma_conexiones_c_8,
    suma_conexiones_c_9, suma_conexiones_c_10, suma_conexiones_c_11, suma_conexiones_c_12,
    suma_conexiones_c_13
]

# Nombres de los distritos
distritos = [
    'AZANGARO', 'CARABAYA', 'CHUCUITO', 'EL COLLAO', 'HUANCANE', 'LAMPA', 'MELGAR', 'MOHO',
    'PUNO', 'SAN ANTONIO DE PUTINA', 'SAN ROMAN', 'SANDIA', 'YUNGUYO'
]

# Total de conexiones de fibra óptica y convencionales en todo el departamento de Puno
total_conexiones_fibra = sum(suma_conexiones_fibra)
total_conexiones_convencionales = sum(suma_conexiones_convencionales)
total_conexiones_departamento = total_conexiones_fibra + total_conexiones_convencionales

# Calcular porcentajes para cada tipo de conexión en cada distrito
porcentajes_fibra = [(f / total_conexiones_departamento) * 100 for f in suma_conexiones_fibra]
porcentajes_convencionales = [(c / total_conexiones_departamento) * 100 for c in suma_conexiones_convencionales]

# Posiciones de las barras en el eje x
distritos_pos = np.arange(len(distritos))

# Ancho de las barras
ancho_barra = 0.65

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))

# Barras de fibra óptica
bars_fibra = plt.bar(distritos_pos, porcentajes_fibra, color='#1f77b4', label='Fibra Óptica', width=ancho_barra)

# Barras de conexiones convencionales
bars_convencionales = plt.bar(distritos_pos + ancho_barra, porcentajes_convencionales, color='#ff7f0e', label='Convencionales', width=ancho_barra)

# Añadir etiquetas de porcentaje a cada barra
for bars in [bars_fibra, bars_convencionales]:
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height:.2f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords='offset points', ha='center', va='bottom', fontsize=8)

# Añadir etiquetas y título
plt.xlabel('Distritos', fontsize=12)
plt.ylabel('Porcentaje de conexiones', fontsize=12)
plt.title('Porcentaje de Conexiones de Fibra Óptica vs. Convencionales por Distrito en Puno', fontsize=14)
plt.xticks(distritos_pos + ancho_barra / 2, distritos, rotation=45, ha='right', fontsize=10)
plt.legend(fontsize=10)

# Quitar bordes
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
